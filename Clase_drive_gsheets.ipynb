## Clase de google sheet


```python
import pandas as pd
import gspread
from google.colab import drive

class ConectorSheet():

  def __init__(self):
    self.ruta = '/content/gdrive'

  def conectarse_drive(self):
    drive.mount(self.ruta)

  def crear_df(self,archivo):
    gc = gspread.service_account(filename=archivo)
    sh_prueba = gc.open('clase_coder').sheet1
    list_of_dicts = sh_prueba.get_all_records()
    df = pd.DataFrame(list_of_dicts)
    return df

  def conectarse_sheet(self,nombre_sheet):
    sh_prueba = gc.open(nombre_sheet).sheet1


```

```python
conn_coder = ConectorSheet()
```

```python
conn_coder.conectarse_drive()
```

```python
%cd gdrive/MyDrive
```

```python
conn_coder.crear_df('clase_coder.json')
```
